```python
# Convert categorical fields into binary fields
df = pd.get_dummies(df)

# Convert all fields into the same data type - float
# df = df.astype('float32')
```

```python
df.drop('Gender_Male', axis = 1, inplace = True)
```

```python
# Separate predictors and the target field
X = df.drop(columns = 'Exited')
y = df['Exited']

# Split data into train, validation and test
# Split all data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 1)
# Further split train into train and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 0.20, random_state = 1)
```

```python
# Create copies of dataframes to avoid indexing warning
[X_train, X_val, X_test, y_train, y_val, y_test] = map(lambda x: x.copy(), [X_train, X_val, X_test, y_train, y_val, y_test])
```

```python
# Scale the Amount field to bring it to the same scale as the rest of the fields
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_val = scaler.transform(X_val)
```

```python
# Creating a function for evaluating model performance
def evaluate_nn(model, train, val, test):
    
    auc_train = roc_auc_score(y_train, model.predict(train))
    auc_val = roc_auc_score(y_val, model.predict(val))
    auc_test = roc_auc_score(y_test, model.predict(test))
    
    return auc_train, auc_val, auc_test
```

#### ReLU Activation

```python
%%time
# Clear the session and set a random seed
tf.keras.backend.clear_session()
tf.random.set_seed(1)

# Initiate the model and add layers
nn_1 = tf.keras.models.Sequential([Input(shape=(X_train.shape[1],)),    # Input layer
                                   Dense(16, activation = 'relu'),      # Hidden layer with relu activation 
                                   Dense(1, activation = 'sigmoid')])   # Output layer

# Add loss function, optimizer and evaluation metric to the model
nn_1.compile(loss = 'binary_crossentropy', 
             optimizer = 'SGD', # Popular optimizers: SGD, Adam, RMSprop
             metrics = ['AUC']) # Many other metrics are avaialble. Eg. accuracy, Recall, Precision, FalseNegatives, etc.

# Train the model
nn_1.fit(X_train, y_train, validation_data = (X_val, y_val), epochs = 10, verbose = 1)

# Score train and test set and record the results
scores = evaluate_nn(nn_1, X_train, X_val, X_test)

# Store results in a dataframe
res = pd.DataFrame({'Method': ['NN - Relu Activation - SGD Optimizer'], 
                    'AUC Train': scores[0], 
                    'AUC Validation': scores[1],
                    'AUC Test': scores[2]})

res
```

# For automated model selection


```python
import h2o
from h2o.automl import H2OAutoML
```

```python
# Start H2O cluster
h2o.init(max_mem_size = 24)

# Shutdown existing cluster
# h2o.cluster().shutdown()
```

```python
# Split all data into train and test sets
train, test = train_test_split(df, test_size = 0.20, random_state = 1)

# Create copies of dataframes to avoid indexing warning
[train, test] = map(lambda x: x.copy(), [train, test])

# Scale the Amount field to bring it to the same scale as the rest of the fields
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
#X_val = scaler.transform(X_val)

# Store predictors in a separate variable
features = [col for col in train.columns if col != 'Exited']
```

```python
# Convert pandas tables into H2O tables
h2o_train = h2o.H2OFrame(train)
h2o_test = h2o.H2OFrame(test)

# Convert target field to categorical to let H2O know that we need a classification model
h2o_train['Exited'] = h2o_train['Exited'].asfactor()
h2o_test['Exited'] = h2o_test['Exited'].asfactor()
```

```python
%%time
# Define and train the model. By default the model will use cross-validation to tune hyperparameters and select best candidate models
h2o_model = H2OAutoML(max_models = 20, max_runtime_secs = 1800, seed = 1)
h2o_model.train(x = features, 
                y = 'Exited', 
                training_frame = h2o_train,  
                leaderboard_frame = h2o_test)
```

```python
# Performance of best models on test set
h2o_model.leaderboard
```

```python
# Create predictions and store in a pandas dataframe
predictions = h2o_model.leader.predict(h2o_test)
predictions = predictions.as_data_frame()
predictions.head()
```

```python
# Compare predictions with actuals
pd.crosstab(y_test, np.where(predictions['p1'] > 0.5, 1,0), colnames = ['Predictions'])
```
